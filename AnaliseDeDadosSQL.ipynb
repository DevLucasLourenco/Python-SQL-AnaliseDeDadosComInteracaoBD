### Importação da Base de Dados

```python
# método1
import pyodbc
import pandas as pd
import locale
locale.setlocale(locale.LC_MONETARY, '')

dados_conexao = ("Driver={SQLite3 ODBC Driver};Server=localhost;Database=salarios.sqlite")

conexao = pyodbc.connect(dados_conexao)

cursor = conexao.cursor()


cursor.execute("""
SELECT * FROM Salaries
               """)
valores = cursor.fetchall()
descricao = cursor.description

cursor.close()
conexao.close()
```

```python
def encontrar_coluna(item):
    return [i[0] for i in item]



tabela_salarios = pd.DataFrame.from_records(valores, columns=encontrar_coluna(descricao))
tabela_salarios
```

### Análise de Dados

```python
tabela_salarios = tabela_salarios.loc[tabela_salarios["Agency"]=="San Francisco", :]
display(tabela_salarios)
print(tabela_salarios.info())
```

##### 1. Qual foi a evolução do salário médio ao longo dos anos

```python
def formatar_moeda(valor):
    return locale.currency(valor, grouping=True)



salarios_media = tabela_salarios.groupby(tabela_salarios['Year']).mean()
del salarios_media['Id']
del salarios_media['BasePay']
del salarios_media['OvertimePay']
del salarios_media['OtherPay']
del salarios_media['Benefits']

salarios_media['TotalPay'] = [formatar_moeda(tp) for tp in salarios_media['TotalPay']]
salarios_media['TotalPayBenefits'] = [formatar_moeda(tpb) for tpb in salarios_media['TotalPayBenefits']]



display(salarios_media)

```

##### 2. Quantos funcionários tivemos ao longo dos anos

```python
aplicar_sep_mihar = lambda x: f"{x:_}".replace('_','.')

quantidade_funcionarios_df = tabela_salarios.groupby(tabela_salarios['Year']).count()
quantidade_funcionarios_df = quantidade_funcionarios_df[['Id']]
quantidade_funcionarios_df = quantidade_funcionarios_df.rename(columns={'Id':'Qtde Funcionários'})
quantidade_funcionarios_df['Qtde Funcionários'] = [aplicar_sep_mihar(s) for s in quantidade_funcionarios_df['Qtde Funcionários']]

quantidade_funcionarios_df
```

##### 3. Qual foi a evolução do total gasto com salário ao longo dos anos

```python
tabela_total = tabela_salarios.groupby(tabela_salarios['Year']).sum()
tabela_total = tabela_total[['TotalPay','TotalPayBenefits']]
tabela_total['TotalPay'] = [formatar_moeda(tp) for tp in tabela_total['TotalPay']]
tabela_total['TotalPayBenefits'] = [formatar_moeda(tp) for tp in tabela_total['TotalPayBenefits']]

display(tabela_total)
```
